```python
from datetime import datetime

from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import requests
```

```python
dataframe = pd.read_csv(
    'longman_list_3000.csv',
    sep=';',
)
dataframe
```

```python
start = datetime.now()

for i in dataframe.index:
    word = dataframe.loc[i, 'word']
    try:
        page = requests.get(f'https://wooordhunt.ru/word/{word}')
        soup = BeautifulSoup(page.text, 'html.parser')
        transcriptions = soup.find_all(class_='transcription')
        if len(transcriptions) == 2:
            dataframe.loc[i, 'pronunciation(US)'] = transcriptions[0].text[2:-1]
            dataframe.loc[i, 'pronunciation(UK)'] = transcriptions[1].text[2:-1]
            print(f'{i + 1} / {dataframe.shape[0]}: {word}')
        elif len(transcriptions) == 4:
            dataframe.loc[i, 'pronunciation(US)'] = f'{transcriptions[0].text[2:-1]} (strong), {transcriptions[1].text[2:-1]} (weak)'
            dataframe.loc[i, 'pronunciation(UK)'] = f'{transcriptions[2].text[2:-1]} (strong), {transcriptions[3].text[2:-1]} (weak)'
            print(f'{i + 1} / {dataframe.shape[0]}: {word}')
        else:
            print(transcriptions)
    except Exception as e:
        print(e)
        continue
        
print(datetime.now() - start)
```
